```python
#Import datetime to allow for conversion

PO1['PO Date']= PO1['PO Date']

#Find the Networkdays merge the orders and find the REC par for Medline Items
Date=PO1.loc[:,'PO Date']
Start_date =str( Date.iloc[1])
End_date =str( Date.iloc[-1])
PO_Date_Start1 = datetime.strptime(Start_date, '%Y-%m-%d')
PO_Date_Start=PO_Date_Start1.date()
PO_Date_End1 = datetime.strptime(End_date, '%Y-%m-%d')
PO_Date_End=PO_Date_End1.date()


days = np.busday_count(PO_Date_Start,PO_Date_End)
# Medline 5 day par calculation
Days=abs(days)/5
Days
```
